```python
import requests
```

### 設定要讀取的網址: pchome 的商品資料

```python
url="http://ecshweb.pchome.com.tw/search/v3.3/all/results?q=sony&page=1&sort=rnk/dc"
```

開始用request get 網頁內容

```python
res = requests.get(url)
```
