```python
import requests
```

### PTT 八卦版網址。然後requests 

```python
url="https://www.ptt.cc/bbs/Gossiping/index.html"
res=requests.get(url)
```

### 第一次進入時就會看到會尋問，是否滿18歲網頁內容

```python
print(res.text)
```

### 查看轉址記錄，因為會被先問轉到18歲的網址去

```python
print(res.history[0])
```

```python
print(res.headers)
```

### 開始加上cooike 傳送

```python
res = requests.get(url, cookies={"over18":"1"})
```

### 列印回傳內容

```python
print(res.text)
```
